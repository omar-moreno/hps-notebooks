```python
#import numpy.lib.recfunctions as rfn

feature_names = [
    'electron_has_l1',
    'positron_has_l1',
    'electron_has_l2',
    'positron_has_l2',
    'positron_d0',
    'positron_phi0',
    'positron_z0', 
    'p_diff'
]

selection = 'positron_has_l1 != -9999'

# Location of the ROOT ntuple
wab_file_path = '/home/omoreno/work/hps/analysis/engrun2015/tridents/mc/wab/trident_analysis.root'
wab_results = rnp.root2array(wab_file_path, 'results', branches=feature_names, selection=selection)
wab_array = rnp.rec2array(wab_results)

wab_positron_has_l1 = wab_results['positron_has_l1']
print len(wab_positron_has_l1)

# Location of the ROOT ntuple
trident_file_path = '/home/omoreno/work/hps/analysis/engrun2015/tridents/mc/tritrig/trident_analysis.root'
trident_results = rnp.root2array(trident_file_path, 'results', branches=feature_names, selection=selection)
trident_results = trident_results[:len(wab_positron_has_l1)]
trident_array = rnp.rec2array(trident_results)

trident_positron_has_l1 = trident_results['positron_has_l1']
print len(trident_positron_has_l1)

# Prepare the targets (class labels) for the training samples
y_trident = np.ones(len(wab_positron_has_l1))
y_wab = np.zeros(len(wab_positron_has_l1))

# Concatenate the WAB and Tri Trig samples
X = np.concatenate((wab_array, trident_array))
#X = X.reshape(len(X), 1)
print "Sample shape: " + str(X.shape)

# Concatenate the WAB and Tri Trig targets
y = np.concatenate((y_wab, y_trident))
print "Targets shape: " + str(y.shape)
```

```python
from sklearn import cross_validation
from sklearn.svm import LinearSVC

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.5, random_state=0)

clf = LinearSVC()

scores = cross_validation.cross_val_score(clf, X.astype(float), y, cv=5)
print scores
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.metrics import confusion_matrix
import itertools


clf = LinearSVC().fit(X_train, y_train)

y_predictions = clf.predict(X_test)
cm = confusion_matrix(y_test, y_predictions)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
cmap=plt.cm.Blues
plt.imshow(cm_normalized, interpolation='nearest', cmap=cmap)
plt.colorbar()
plt.tight_layout()

thresh = cm_normalized.max() / 2.
for i, j in itertools.product(range(cm_normalized.shape[0]), range(cm_normalized.shape[1])):
        plt.text(j, i, round(cm_normalized[i, j], 2),
                 horizontalalignment="center",
                 color="white" if cm_normalized[i, j] > thresh else "black")

tick_marks = np.arange(len(['wab', 'trident']))
plt.xticks(tick_marks, ['wab', 'trident'])
plt.yticks(tick_marks, ['wab', 'trident'])
plt.ylabel('True label')
plt.xlabel('Predicted label');
```
